```python
# Essential libraries for data analysis
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Visualization settings
sns.set(style="whitegrid")
plt.rcParams["figure.figsize"] = (10, 6)

# Load datasets
happiness_df = pd.read_excel('../data/WorldHappinessReport-2024.xlsx')
oecd_filtered_df = pd.read_csv('../data/OECD,DF_BLI,+all_Filtered_Data.csv')
oecd_unfiltered_df = pd.read_csv('../data/OECD,DF_BLI,+all_Unfiltered_Data.csv')

# Preview datasets
print("World Happiness Report Dataset:")
display(happiness_df.head())

print("OECD Better Life Index (Filtered) Dataset:")
display(oecd_filtered_df.head())

print("OECD Better Life Index (Unfiltered) Dataset:")
display(oecd_unfiltered_df.head())
```

## Step 3: Exploratory Data Analysis (EDA)

### World Happiness Report (happiness_df)

```python
# Structure and data types
happiness_df.info()

# Count missing values
missing_values = happiness_df.isnull().sum().sort_values(ascending=False)
print("Missing Values per Column:\n")
print(missing_values)
```

```python
# Basic stats for numeric columns
happiness_df.describe()
```

```python
# Distribution of Ladder Scores
sns.histplot(happiness_df['Ladder score'], bins=30, kde=True)
plt.title("Distribution of Global Happiness Scores")
plt.xlabel("Ladder Score")
plt.ylabel("Frequency")
plt.show()
```

```python
# Average happiness score per year
avg_by_year = happiness_df.groupby("Year")["Ladder score"].mean().reset_index()

sns.lineplot(data=avg_by_year, x="Year", y="Ladder score", marker="o")
plt.title("Average Global Happiness Score Over Time")
plt.xlabel("Year")
plt.ylabel("Average Ladder Score")
plt.show()
```

## Step 4: Data Cleaning & Preparation

### 4.1: Assess Missingness in Key Columns

```python
# Select only the relevant explanatory variables
explanatory_cols = [
    'Explained by: Log GDP per capita',
    'Explained by: Social support',
    'Explained by: Healthy life expectancy',
    'Explained by: Freedom to make life choices',
    'Explained by: Generosity',
    'Explained by: Perceptions of corruption'
]

# Check how many rows are fully complete
complete_rows = happiness_df.dropna(subset=explanatory_cols)
print(f"Complete rows: {len(complete_rows)} / {len(happiness_df)}")
```

### 4.2 Create a Cleaned Dataset

```python
# Keep only necessary columns
columns_to_keep = ['Year', 'Country name', 'Ladder score'] + explanatory_cols
happiness_clean = happiness_df[columns_to_keep].dropna()

# Check new shape
print("Cleaned dataset shape:", happiness_clean.shape)

# Preview cleaned data
happiness_clean.head()
```

## Step 5: Correlation & Relationship Analysis
We'll use the cleaned dataset happiness_clean to uncover relationships between the happiness score (Ladder score) and its explanatory variables.



### 5.1: Correlation Matrix

```python
# Compute correlation matrix
corr_matrix = happiness_clean.corr(numeric_only=True)

# Visualize as a heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title("Correlation Matrix: Happiness and Explanatory Variables")
plt.show()
```

Which variables have the strongest correlation (positive or negative) with Ladder score? For example, GDP and social support are usually top contributors.



### 5.2: Pairplot to Visualize Individual Relationships

```python
# Smaller subset for pairplot
sns.pairplot(happiness_clean[[
    'Ladder score',
    'Explained by: Log GDP per capita',
    'Explained by: Social support',
    'Explained by: Healthy life expectancy'
]])
plt.suptitle("Pairwise Relationships Between Key Variables", y=1.02)
plt.show()
```

### 5.3: Highlight Top Correlators with Happiness

```python
# Extract correlations with Ladder Score
ladder_corr = corr_matrix['Ladder score'].sort_values(ascending=False)

print("Top correlations with happiness score:")
print(ladder_corr)
```

## Step 6: Advanced Analysis – Linear Regression

Use statsmodels to run an Ordinary Least Squares (OLS) regression. This will help us:

Understand which features significantly affect happiness

Quantify their effect sizes

Assess model fit (via R²)



### 6.1: Define the Model

```python
import statsmodels.api as sm

# Define features (X) and target (y)
X = happiness_clean[[
    'Explained by: Log GDP per capita',
    'Explained by: Social support',
    'Explained by: Healthy life expectancy',
    'Explained by: Freedom to make life choices',
    'Explained by: Generosity',
    'Explained by: Perceptions of corruption'
]]
y = happiness_clean['Ladder score']

# Add a constant term for intercept
X = sm.add_constant(X)
```

### 6.2: Fit the Model and View Results

```python
# Fit the regression model
model = sm.OLS(y, X).fit()

# Show summary
model.summary()
```

R-squared: How much of the variation in happiness the model explains.

coef: Size and direction of each variable’s influence.

P>|t|: Statistical significance (values < 0.05 are usually significant).

Confidence intervals: Range in which the true coefficient likely lies.

### 6.3: Optional – Residual Analysis (Check Model Validity)

```python
# Plot residuals to check for patterns
residuals = model.resid

sns.histplot(residuals, bins=30, kde=True)
plt.title("Distribution of Regression Residuals")
plt.xlabel("Residual")
plt.show()
```

##  Step 7: Insights & Conclusions

### 7.1: Key Findings

## Key Insights

- **GDP per capita** and **Social support** are the strongest predictors of happiness based on correlation and regression analysis.
- **Freedom to make life choices** and **Healthy life expectancy** also have strong, statistically significant effects.
- **Generosity** and **Perceptions of corruption** show weaker or inconsistent influence across countries.
- The **average global happiness score** has remained relatively stable over time, with top countries like Finland consistently scoring highest.
- Countries with low happiness tend to suffer from weak social support, low GDP, and governance issues.

## Regression Summary Interpretation

- The model explains ~XX% (your R² value) of the variation in happiness, which is solid for real-world data.
- Variables with statistically significant p-values (< 0.05) should be viewed as reliable contributors to happiness.
- Residuals are reasonably normally distributed, suggesting the model assumptions are well met.

## Limitations

- Missing data led to dropping a number of rows, potentially biasing results toward better-documented countries.
- Data is observational, not causal — correlations ≠ causation.
- Cultural and contextual factors not in the data (e.g., values, religion, climate) might play a big role.

## Summary

This analysis gives a data-driven view into what makes nations happier. It reinforces the importance of not only economic prosperity but also social and civic factors. These insights can guide policymakers, researchers, and individuals toward more effective decisions for improving well-being.
